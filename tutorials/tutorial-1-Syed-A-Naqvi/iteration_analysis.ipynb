This notebook is going to be used to generate visual representations by way of graphs
so that we can better evaluate the behaviour of the iterative rule:
phi(x)=(1/2)*(x+(a/x)) for various error tolerances (epsilon_x)

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#This is the logic of the iteration function without any print statements to simplify the analysis process

def iteration(x, a, k_max, eps_x):

    for k in range(k_max):
        x_previous = x              #This line keeps track of the previous iterate
        x = phi(x,a)
        
        if( abs(x-x_previous) < eps_x ):
            return x, k+1

def phi(x, a):
    return (1/2)*(x+(a/x))
```

```python
x=[]
y=[]

for i in range(-1000,1000):
    if i in [0]:
        continue
    
    na,iterations = iteration(i,5,50,1.0e-4)
    x.append(i)
    y.append(iterations)

#print(x)
#print(y)
plt.plot(x,y)

plt.xlabel("Initial Values")
plt.ylabel("Number of Iterations")
plt.title('Number of Iterations Based on Initial Value')
plt.show()
```

From the above plot it becomes readily apparent that all values will eventually converege to below the error tolerance defined. However, as the initial values move farther away from 0 in the negative and positive directions, the number of iterations required for convergence increases. 

```python
x=[]
y=[]

for i in range(0,-20,-2):
    
    na,iterations = iteration(3,5,100,1.0*10**i)
    x.append(str(1.0*10**i))
    y.append(iterations)

print(x)
print(y)
plt.plot(y,x)

plt.ylabel("Epsilon")
plt.xlabel("Number of Iterations")
plt.title('Number of Iterations Based on Error')
plt.show()
```

Based on the above visual, keeping everything the same while varying the error threshold results in a relationship that appears roughly exponential. We can see the the number of iterations it takes for the function to converge increases exponentially as we decrease the permissable error threshold.
